```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score,classification_report
from sklearn.feature_selection import SelectKBest,f_classif
```

```python
df=pd.read_csv('CarPrice_dataset.csv')
df
```

```python
df.info()
```

```python
print(df.columns)
```

```python
df.describe(include='all')    # getting descriptive stats
```

```python
# dropping the CarName column

df.drop('CarName',axis=1,inplace=True)
df.describe(include='all')
```

```python
# checking missing values
df.isnull().sum()
```

```python
# Converting categorical data columns to Numerical using LabelEncoder

le = LabelEncoder()
for column in df.select_dtypes(include=['object']).columns:
    df[column] = le.fit_transform(df[column])
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Splitting the data into training and test data

x= df.drop(['price'],axis=1)
y = df['price']
```

```python
print(x)
```

```python
# splitting training and test data

x_train,x_test,y_train,y_test =train_test_split(x,y, test_size = 0.2,random_state = 42)
```

1.Linear Regression

```python
# Model Implementation

linreg_model = LinearRegression()  #Loading linearRegression model

linreg_model.fit(x_train,y_train)



```

```python
# predictions
y_pred = linreg_model.predict(x_test)

```

```python
# Evaluation 
mse = mean_squared_error(y_test,y_pred)      
print(f'Mean Squared Error: {mse}')

mae = mean_absolute_error(y_test,y_pred)
print(f'Mean Absolute Error: {mae}')

r2 = r2_score(y_test,y_pred)
print(f'R-squared: {r2}')
```

2 . Decision Tree Regressor

```python
# Model Implementation

des_tr_model = DecisionTreeRegressor(random_state = 42)  #Loading DecisionTreeRegressor model

des_tr_model.fit(x_train,y_train)



```

```python
# predictions
y_pred = des_tr_model.predict(x_test)

```

```python
# Evaluation 
mse = mean_squared_error(y_test,y_pred)      
print(f'Mean Squared Error: {mse}')

mae = mean_absolute_error(y_test,y_pred)
print(f'Mean Absolute Error: {mae}')

r2 = r2_score(y_test,y_pred)
print(f'R-squared: {r2}')
```

3. Random Forest regression

```python
# Model Implementation

rand_for_model = RandomForestRegressor(random_state = 42)  #Loading RandomForestRegression model

rand_for_model.fit(x_train,y_train)



```

```python
# predictions
y_pred = rand_for_model.predict(x_test)

```

```python
# Evaluation 
mse = mean_squared_error(y_test,y_pred)      
print(f'Mean Squared Error: {mse}')

mae = mean_absolute_error(y_test,y_pred)
print(f'Mean Absolute Error: {mae}')

r2 = r2_score(y_test,y_pred)
print(f'R-squared: {r2}')
```

4.Support Vector Regression

```python
# Model Implementation

svr_model = SVR(kernel='linear')  #Loading SVR model

svr_model.fit(x_train,y_train)
```

```python
# predictions
y_pred = svr_model.predict(x_test)

```

```python
# Evaluation 
mse = mean_squared_error(y_test,y_pred)      
print(f'Mean Squared Error: {mse}')

mae = mean_absolute_error(y_test,y_pred)
print(f'Mean Absolute Error: {mae}')

r2 = r2_score(y_test,y_pred)
print(f'R-squared: {r2}')
```

5. Gradient Boosting regression

```python
# Model Implementation

Gr_boost_model = GradientBoostingRegressor(random_state = 42)  #Loading SVR model

Gr_boost_model.fit(x_train,y_train)
```

```python
# predictions
y_pred =Gr_boost_model.predict(x_test)

```

```python
# Evaluation 
mse = mean_squared_error(y_test,y_pred)      
print(f'Mean Squared Error: {mse}')

mae = mean_absolute_error(y_test,y_pred)
print(f'Mean Absolute Error: {mae}')

r2 = r2_score(y_test,y_pred)
print(f'R-squared: {r2}')
```

 Hyperparameter Tuning

```python
# Regression model (Linear Regression) with hyperparameter tuning
param_grid_regression = {'fit_intercept': [True, False], 'positive': [True, False]}
grid_search_regression = GridSearchCV(LinearRegression(), param_grid=param_grid_regression,cv =5)
grid_search_regression.fit(x_train, y_train)
best_params_regression = grid_search_regression.best_params_
best_model_regression = grid_search_regression.best_estimator_
# Make predictions
y_pred_regression = best_model_regression.predict(x_test)
mse_regression = mean_squared_error(y_test, y_pred_regression)
print("Regression MSE:", mse_regression)
```

```python
# Regression model (DecisonTree Regression) with hyperparameter tuning
param_grid_regression = {'max_depth': [True, False], 'max_features': [True, False]}
grid_search_regression = GridSearchCV(DecisionTreeRegressor(), param_grid=param_grid_regression,cv =5)
grid_search_regression.fit(x_train, y_train)
best_params_regression = grid_search_regression.best_params_
best_model_regression = grid_search_regression.best_estimator_
# Make predictions
y_pred_regression = best_model_regression.predict(x_test)
mse_regression = mean_squared_error(y_test, y_pred_regression)
print("Regression MSE:", mse_regression)
```

```python
# Regression model (Support Vector Regression) with hyperparameter tuning
param_grid_regression = {'degree': [True, False], 'gamma': [True, False]}
grid_search_regression = GridSearchCV(SVR(), param_grid=param_grid_regression,cv =5)
grid_search_regression.fit(x_train, y_train)
best_params_regression = grid_search_regression.best_params_
best_model_regression = grid_search_regression.best_estimator_
# Make predictions
y_pred_regression = best_model_regression.predict(x_test)
mse_regression = mean_squared_error(y_test, y_pred_regression)
print("Regression MSE:", mse_regression)
```

```python
# Regression model (Random Forest Regression) with hyperparameter tuning
param_grid_regression = {'max_depth': [True, False], 'max_features': [True, False]}
grid_search_regression = GridSearchCV(RandomForestRegressor(), param_grid=param_grid_regression,cv =5)
grid_search_regression.fit(x_train, y_train)
best_params_regression = grid_search_regression.best_params_
best_model_regression = grid_search_regression.best_estimator_
# Make predictions
y_pred_regression = best_model_regression.predict(x_test)
mse_regression = mean_squared_error(y_test, y_pred_regression)
print("Regression MSE:", mse_regression)
```
